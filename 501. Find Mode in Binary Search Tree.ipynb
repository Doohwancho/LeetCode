```python
"""
Given a binary search tree (BST) with duplicates, find all the mode(s) (the most frequently occurred element) in the given BST.

Assume a BST is defined as follows:

The left subtree of a node contains only nodes with keys less than or equal to the node's key.
The right subtree of a node contains only nodes with keys greater than or equal to the node's key.
Both the left and right subtrees must also be binary search trees.
 

For example:
Given BST [1,null,2,2],

   1
    \
     2
    /
   2
 

return [2].

Note: If a tree has more than one mode, you can return them in any order.
"""

#문제

#트리 안에 가장 빈도수가 많은 값을 리스트에 담아서 반환하라.
#만약 모든 노드값의 빈도수가 같다면, 모든 노드값을 반환하고, 트리가 비어있으면 []를 반환하라.


#Trial01

#dfs로 트리를 순회하면서, 리스트에 담은 값들을 max(set(list), key=list.count) 을 이용하여 
#가장 빈도수가 많은 element를 도출하는데 까지는 성공했지만, 만약 가장 빈도수가 많은 수가 없이 모두 동일한 경우
#(e.g 모든 숫자가 각각 한개씩 있다면) 가장 첫번째에 있는 element만 나오기 때문에,
#mode를 도출하는 더 좋은 방법을 찾아야 함


# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def findMode(self, root: TreeNode) -> List[int]:
        self.mode = []
        def dfs(root):
            if not root: return
            self.mode.append(root.val)
            if root.left: dfs(root.left)
            if root.right: dfs(root.right)
        
        dfs(root)
        return [max(set(self.mode), key=self.mode.count)] if self.mode else []
    
    
#Solution by gabbu

#문제풀이

#먼저 collections.defaultdict 으로 빈 딕셔너리를 만든다. 트리를 순회하면서 key값을 해당 노드의 값, value값을
#해당 노드의 값의 빈도수(같은값이 나올때마다 +1)을 한다.
#max()를 사용하여 딕셔너리의 value값 중 가장 높은 숫자(최다빈도)를 구하여 maxMode에 넣고
#for문을 돌리면서 value의 값이 maxMode와 같은 값을 리스트에 담아 리턴한다.
#만약 모든 element의 빈도수가 같아도(ex.1이라도) 괜찮다.
#이 경우 모든 element의 빈도수가 maxMode인 1과 같으므로 모든 element을 반환시켜 주기 때문이다.



from collections import defaultdict

class Solution(object):
    def dfs(self, root, dictionary):
        if not root: return
        dictionary[root.val]+=1
        if root.left: self.dfs(root.left, dictionary)
        if root.right: self.dfs(root.right, dictionary)
    

    def findMode(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root: return []
        dictionary = defaultdict(int)
        self.dfs(root,dictionary)
        maxMode = max(dictionary.values())
        result = [k for k,v in dictionary.items() if v == maxMode]
        return result
    
```
