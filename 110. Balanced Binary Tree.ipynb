```python
"""
Given a binary tree, determine if it is height-balanced.

For this problem, a height-balanced binary tree is defined as:

a binary tree in which the depth of the two subtrees of every node never differ by more than 1.

Example 1:

Given the following tree [3,9,20,null,null,15,7]:

    3
   / \
  9  20
    /  \
   15   7
Return true.

Example 2:

Given the following tree [1,2,2,3,3,null,null,4,4]:

       1
      / \
     2   2
    / \
   3   3
  / \
 4   4
Return false.
"""

#문제

#subtree의 depth(or level)이 1 이상 차이나면 False를 반환아고, 아니면 True를 반환하라



#문제풀이

#depth를 구하는 함수를 만든 후, left child와 right child의 depth를 각각 구해서 그 차이가 2보다 작으면 True를 반환한다.



# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isBalanced(self, root: TreeNode) -> bool:
        def getNodeDepth(node):
            if(node==None):return 0;

            queue=[(node,1)];
            depth=1;

            while(len(queue)):
                first,depth=queue.pop(0);
                if(first.left!=None):
                    queue.append((first.left,depth+1));
                if(first.right!=None):
                    queue.append((first.right,depth+1));
            return depth;

        if(root==None):return True;
        l=getNodeDepth(root.left);
        r=getNodeDepth(root.right);

        return (abs(l-r)<2) and self.isBalanced(root.left) and self.isBalanced(root.right);
```
