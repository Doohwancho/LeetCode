Given an array of 4 digits, return the largest 24 hour time that can be made.

The smallest 24 hour time is 00:00, and the largest is 23:59.  Starting from 00:00, a time is larger if more time has elapsed since midnight.

Return the answer as a string of length 5.  If no valid time can be made, return an empty string.

 

>Example 1:<br>
>Input: [1,2,3,4]<br>
>Output: "23:41"<br>

>Example 2:<br>
>Input: [5,5,5,5]<br>
>Output: ""<br>

```python
#문제
#4개의 숫자가 들어있는 리스트를 주면, 시간단위로 조합해서 가장 높은(?) 시간(24:00과 가까운) 리턴하라.
#예를들어 [1,2,3,4]가 주워졌으면,
#12:34 / 21:34 / 23:41 등등의 경우의 수 중에서, 가장 높은(?)시간인 23:41을 리턴한다.



#Trial01

#일단 4개 숫자중, 가장 높은 시간을 먼저 구하고, 나머지 숫자를 조합하여 가장 높은 분을 구하는 방법을 썼다.
#하지만 [2,0,6,6]에서 막힌다.
#20시가 완성되지만, 66으로 유효한 분을 만들 수 없기 때문이다.



# A = [4,2,3,1]
# A = [1,2,3,4]
# A = [5,5,5,5]
# A = [9,0,7,7]
# A = [0,0,0,2]
A = [2,0,6,6]

hour=[]
minute = []

#max hour
for i in range(len(A)):
    for j in range(i+1,len(A)):
        if 24 > int(str(A[i])+str(A[j])): 
            hour.append(str(A[i])+str(A[j]))
        if 24 > int(str(A[j])+str(A[i])):
            hour.append(str(A[j])+str(A[i]))
            

#remove duplicates
if hour:
    for k in max(hour):
        A.remove(int(k))
else:
    print("None hour")
    

if 60 > int(str(A[0])+str(A[1])):
    minute.append(str(A[0])+str(A[1]))
if 60 > int(str(A[1])+str(A[0])):
    minute.append(str(A[1])+str(A[0]))
    
if minute:    
    print(max(hour)+':'+max(minute))
else:
    print("None minute")
```

```python
#Trial 02 -- SUCCEED

#step01 - 4개 숫자의 모든 경우의 수를 구한다. (permutation)
#step02 - 구할때 첫번째 자리 숫자는 2 이하, 두번째 자리 숫자는 첫번째 자리 숫자가 2일때 3 이하, 세번째 자리 숫자는 5 이하의 조건을 건다.



# permutation 구하는 함수. 너무 어려우니 세부 로직은 pass한다. 
#출처 : https://www.geeksforgeeks.org/permutation-and-combination-in-python/
def permutation(lst): 
  
    # If lst is empty then there are no permutations 
    if len(lst) == 0: 
        return [] 
  
    # If there is only one element in lst then, only 
    # one permuatation is possible 
    if len(lst) == 1: 
        return [lst] 
  
    # Find the permutations for lst if there are 
    # more than 1 characters 
  
    l = [] # empty list that will store current permutation 
  
    # Iterate the input(lst) and calculate the permutation 
    for i in range(len(lst)): 
        m = lst[i]
  
        # Extract lst[i] or m from the list.  remLst is 
        # remaining list 
        remLst = lst[:i] + lst[i+1:] 

        # Generating all permutations where m is first 
        # element 
        for p in permutation(remLst): 
            l.append([m] + p) 
    return l




A = [2,0,6,6]
#A = [5,5,5,5]
rst = []

#permutation으로 A의 모든 경우의 수를 구한다.
for p in permutation(A): 
    if (p[0] <=  2 and p[2] < 6):        #시간은 23시가 최대, 분은 59분이 최대이기 때문에, 앞자리 조건문을 걸어준다.
        rst.append(p)
        
        
#print(rst)
final_rst = rst.copy()

for i in rst:
    if i[0] == 2 and i[1] >= 4:
        final_rst.remove(i)               #시간의 앞자리수가 2일경우, 24시,25시,26시...가 나올 수 없으므로 조건문을 걸어준다

if final_rst:
    time = ''.join([str(i) for i in max(final_rst)])
    print(time[:2]+':'+time[2:])
else:
    print("")
```
