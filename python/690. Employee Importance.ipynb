You are given a data structure of employee information, which includes the employee's unique id, his importance value and his direct subordinates' id.

For example, employee 1 is the leader of employee 2, and employee 2 is the leader of employee 3. They have importance value 15, 10 and 5, respectively. Then employee 1 has a data structure like [1, 15, [2]], and employee 2 has [2, 10, [3]], and employee 3 has [3, 5, []]. Note that although employee 3 is also a subordinate of employee 1, the relationship is not direct.

Now given the employee information of a company, and an employee id, you need to return the total importance value of this employee and all his subordinates.

>Example 1:<br>
Input: [[1, 5, [2, 3]], [2, 3, []], [3, 3, []]], 1<br>
Output: 11<br>
Explanation:<br>
>Employee 1 has importance value 5, and he has two direct subordinates: employee 2 and employee 3. They both have importance value 3. So the total importance value of employee 1 is 5 + 3 + 3 = 11.<br>

```python
#문제

#리스트가 [[1, 5, [2, 3]], [2, 3, []], [3, 3, []]] 이런식으로 주어진다. RDBMS라고 이해하면 편하다.
#RDBMS는 관계형 데이터 베이스라는 뜻으로, 쉽게 말해 데이터 마다 서로 연결되어 있다는 뜻이다.
#리스트를 element단위로 보면, 첫번째 리스트는 [1, 5, [2, 3]] 이렇게 생겼다. 
#첫번째 숫자 1은 사원 번호이다. 사원번호가 1에 가까울수록 상사이다.
#두번째 숫자 5는 사원의 가치이다. 이 문제는 총 사원의 가치를 구하는 문제이니 이 숫자가 중요하다.
#세번째는 리스트로써 [2,3]이라고 나온다. 여기서 [2,3]의 의미는, 사원번호 1의 종속된 부하직원이 2랑 3이라는 말이다.
#우리는 사원번호를 넣으면, 해당사원과 그 사원의 종속된 부하직원들의 가치의 총 합을 구하여야 한다.
#만약 부하직원이 또다른 부하직원이 있다면,
#예를들어, 첫번째 사원이 [1, 5, [2, 3]]이고,
#두번째 사원이 [2, 3, [4]]이고,
#네번째 사원이 [4,2,[]]]이면, 네번째 사원도 두번째 사원에 종속되기 때문에, 가치의 총 합을 더할 때, 네번째 사원의 가치도 더해주어야 한다.



#문제 풀이

#한 사람을 기준으로 부하직원들이 트리 형식으로 되어있다.
#문제 접근 방식은, 가장 첫번째의 가장 마지막에 있는 부하직원의 value를 구하면 그 부하직원을 지워주고, 점점 위로 올라가는
#down-top 방식을 사용했다.

#이해를 돕기 위해 코드를 시각화 해주는 사이트를 첨부한다.
#http://pythontutor.com/live.html#mode=edit





#문제 결함

#leetcode에서 답 제출시 input을 [[1,2,[2]], [2,3,[]]], 2 이런식으로 준다. 
#employee들이 한 리스트에 담겨져 있지 않다. input이 이상하다.
#따라서 밑의 풀이는 leetcode상에는 pass 되지 않고, 공부 목적으로 업로드 한다.


employees = [[1, 5, [2, 3]], [2, 3, [4]], [3, 3, []], [4,2,[]]], 1

#주어진 employees에서 전체사원정보와 타깃하는 사원을 분리한다.
valueList, target = employees

#전체 사원정보를 딕셔너리 형태로 만든다.
dictionary = { a: [b,c] for a,b,c in valueList }

#총 가치를 담을 변수를 선언한다.
value = 0

#만약 가장 마지막에 있는 부하직원의 가치를 더해주면, 해당 직원의 정보를 지우기 위해 remove를 boolean형태로 선언한다.
remover = False

#가장 처음 상사의 id를 저장해둔다.
parent = target

while dictionary[target][1]:

    #상사와 부하직원을 왔다갔다 할 것이기 때문에, 상사의 id를 pre, 부하직원의 id를 pro로 선언한다.
    pre = target
    
    #가장 마지막 직원의 value를 더했으면, 해당 직원의 정보를 없애준다
    if remover:
        dictionary[target][1].pop(0)
        remover = False
    
    #만약 상사가 부하직원이 있다면
    if dictionary[pre][1]:
        #먼저 상사의 부하직원들 중 첫번째 사람의 id를 pro 변수에 저장한다.
        pro = dictionary[pre][1][0]
        target = pro
        #만약 부하직원의 종속된 부하직원이 없다면 (= 마지막 부하직원이라면)
        if not dictionary[pro][1]:
            #마지막 부하직원의 가치를 더해주고
            value += dictionary[pro][0]
            #상사id로 되돌아 간 다음
            target = pre
            #해당 부하직원의 정보를 삭제한다
            remover = True
    else:
        #만약 상사가 부하직원이 없다면, 최고참 상사로 돌아간다.
        target = parent
        #만약 최고참 상사의 부하직원의 가치를 이미 모두 더해서, 남은 부하직원의 정보가 없다면, 해당 상사의 value를 더해주고 함수는 마무리된다.
        if not dictionary[target][1]:
            value += dictionary[target][0]
        
print(value)
```
