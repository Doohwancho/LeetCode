```python
"""
Given a binary tree, check whether it is a mirror of itself (ie, symmetric around its center).

For example, this binary tree [1,2,2,3,4,4,3] is symmetric:

    1
   / \
  2   2
 / \ / \
3  4 4  3
 

But the following [1,2,2,null,3,null,3] is not:

    1
   / \
  2   2
   \   \
   3    3
"""

#문제

#주어진 트리가 좌우대칭이면 True, 아니면 False를 반환하라


#Trial01 
#???  -  return root.left == root.right 부분에서 막힌다.

#방법: right child를 왼쪽 값&오른쪽 값을 바꿔준 다음, 기존의 left child와 뒤집혀진 right child가 같은지 비교한다.
#error: right child를 뒤집는데는 성공했으나, 맨 마지막 left child와의 비교 부분에서 막힌다. 왜지??? 


# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isSymmetric(self, root: TreeNode) -> bool:
#         def dfs(root):
#             if not root: return
#             if root.left: dfs(root.left)
#             print(root.val)
#             if root.right: dfs(root.right)
        
        def reverser(root):
            if not root: return
            if root.left and root.right:
                tmp = None
                tmp = root.right.val
                root.right.val = root.left.val
                root.left.val = tmp
                reverser(root.left)
                reverser(root.right)
            elif root.left:
                root.right = root.left
                root.left = None
                reverser(root.right)
            elif root.right:
                root.left = root.right
                root.right = None
                reverser(root.left)
            else:
                return
        
        if root:
            if root.left or root.right:
                #print("original")
                #dfs(root.right)
                reverser(root.right)
                #print("left")
                #dfs(root.left)
                #print("right")
                #dfs(root.right)
                return root.left == root.right
        else:
            return False
        
        
#Solution by realisking

#left child와 right child를 교차해서 넣어 재귀용법으로 비교하되, 
#주의할 사항은 left child==right child를 left child와 right child가 모두 있을 때만 해야 한다.

class Solution(object):
    def isSymmetric(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        def helper(L, R):
            return L.val == R.val and helper(L.left, R.right) and \
                   helper(L.right, R.left) if L and R else L == R == None
        return helper(root, root)
```
