You are given a map in form of a two-dimensional integer grid where 1 represents land and 0 represents water.

Grid cells are connected horizontally/vertically (not diagonally). The grid is completely surrounded by water, and there is exactly one island (i.e., one or more connected land cells).

The island doesn't have "lakes" (water inside that isn't connected to the water around the island). One cell is a square with side length 1. The grid is rectangular, width and height don't exceed 100. Determine the perimeter of the island.

 

>Example:<br>
><br>
>Input:<br>
>[[0,1,0,0],<br>
> [1,1,1,0],<br>
> [0,1,0,0],<br>
> [1,1,0,0]]<br>
><br>
>Output: 16<br>
><br>
>Explanation: The perimeter is the 16 yellow stripes in the image below:<br>

```python
#문제

#[[0,1,0,0],
#[1,1,1,0],
#[0,1,0,0],
#[1,1,0,0]]

#matrix가 주어진다. 1은 섬이고 0은 바다다. 섬의 총 둘레를 구하라.




#문제풀이

#섬은 사각형이니 4면이므로 둘레는 4다. 
#그런데 섬들이 붙어있을때 둘레를 구하려면, 나란히 붙어있는것을 고려해야한다.

#예를들어, 섬 두개가 나란히 붙어있는 것의 둘레는 6이다. 그런데 이를 구하려면
#단순히 4*2만 해서 되는게 아니라, 겹치는 한 면도 고려해서 
#4*2(4면씩 두개의 섬 == 8) - 1(섬끼리 붙어있는 횟수)*2 하면 6이 나온다.

#섬의 갯수를 구하는건 단순히 리스트 안에 1의 갯수를 세면 된다.
#섬 두끼리 겹치는 것은 크게 두가지 방법을 써서, 1) 가로 겹치는 것만 카운트 2) 세로 겹치는 것만 카운트 하고 서로 합한다.


#step01 - island의 수를 구한다
#step02 - island끼리 겹치는 수를 구한다
#step03 - island의 둘레는, (전체 island수 * 4) - (겹치는 수 * 2) 이므로, 식을 적용한다.


def islandPerimeter(grid):
    
    #island의 갯수
    numisland = 0

    #island가 겹치는 수
    cnt = 0

    #step01 - island의 갯수 구하기
    for m in grid:
        numisland += m.count(1) #해당 행 리스트에 1의 숫자만큼 섬의 수를 더한다.



    #step02 - island끼리 겹치는 수를 구하기

    #island의 행(가로)를 for문으로 돌면서, 연속된 1이 있나(겹치는 island가 있나) 확인
    for i in range(len(grid)):
        #첫번째부터 다음것을 비교하면 맨 마지막에서 비교대상이 없어서 에러나기 때문에 두번째부터 그 이전것을 비교
        for j in range(1, len(grid[i])): 
            if grid[i][j] == grid[i][j-1] == 1:  #여기서 행들의 가로를 비교하게 설정한다.
                cnt += 1
        print("row finished", cnt)


    #만약 리스트가 [[1,0]]처럼 주워져서, 세로로 비교할게 없다면 pass
    if len(grid) > 1:
        #island의 행(세로)를 for문으로 돌면서, 연속된 1이 있나(겹치는 island가 있나) 확인
        for k in range(1, len(grid)):
            for l in range(len(grid[k])):
                if grid[k][l] == grid[k-1][l] == 1: #여기서 행들의 세로를 비교하게 설정한다
                    cnt += 1
            print("column finished", cnt)
    else:
        pass


    #step03 
    return(numisland * 4 - cnt * 2)





grid = [
        [0,1,0,0],
        [1,1,1,0],
        [0,1,0,0],
        [1,1,0,0]
        ]

# grid = [
#         [1,0]
#         ]

# grid = [
#         [1,0,0],
#         [0,0,0]
#         ]
    
islandPerimeter(grid)
```
