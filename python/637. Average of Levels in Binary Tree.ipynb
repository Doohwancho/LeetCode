```python
"""
Given a non-empty binary tree, return the average value of the nodes on each level in the form of an array.
Example 1:
Input:
    3
   / \
  9  20
    /  \
   15   7
Output: [3, 14.5, 11]
Explanation:
The average value of nodes on level 0 is 3,  on level 1 is 14.5, and on level 2 is 11. Hence return [3, 14.5, 11].
"""

#문제
#각층의 값들의 평균을 구하여 리스트에 넣어라


#문제풀이
#각층의 값들을 구하는 문제와 유사하다. (https://doohwancho.tistory.com/90)
#먼저 각 층의 값들을 구하여 리스트에 담아놓은 후,
#각 리스트의 값들을 평균어 답을 구한다.




# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def averageOfLevels(self, root: TreeNode) -> List[float]:
        q, level = [root], []                         #q와 rst변수를 만들어준다. level는 각층의 값을 넣어줄 리스트다. 
        if root:                                      #root가 None일 경우 [] return
            while q:                                  #while문으로 q를돈다. q는 매 level마다 그 아래 children level의 값들로 교체된다.
                level.append([node.val for node in q])#해당 level에 값들을 넣어준다
                tmp = []
                for node in q:                        #아래 children level의 값들을 넣어준다.
                    if node.left != None:             #children의 값이 없을 수도 있으므로, 경우의 수를 고려하는 코드를 넣는다.
                        tmp.append(node.left)
                    if node.right != None:
                        tmp.append(node.right)
                q = tmp                               #아래 level의 값들을 넣어준 tmp를 q에 넣어주고, 다시 while문을 돌린다.
            
            ans=[]                                    #모든층을 돌면서 값을 넣는게 완료되면, 각 층의 값들을 평균시키고 넣어줄 리스트를 만든다.
            
            for i in level:                           #각 층마다 돌면서 평균내주는 코드
                avg = 0
                for j in i:
                    avg += j
                ans.append(avg/len(i))
                
            return ans
        else:
            return []
```
