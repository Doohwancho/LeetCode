```python
"""
Given two non-empty binary trees s and t, check whether tree t has exactly the same structure and node values with a subtree of s. A subtree of s is a tree consists of a node in s and all of this node's descendants. The tree s could also be considered as a subtree of itself.

Example 1:
Given tree s:

     3
    / \
   4   5
  / \
 1   2
Given tree t:
   4 
  / \
 1   2
Return true, because t has the same structure and node values with a subtree of s.
Example 2:
Given tree s:

     3
    / \
   4   5
  / \
 1   2
    /
   0
Given tree t:
   4
  / \
 1   2
Return false.
"""
#문제

#트리 s와 서브트리 t가 주어지면, 트리 s안의 서브트리중, t와 같은 서브트리가 있으면 True, 아니면 False를 반환하라.


#문제풀이

#preorder traverse 방식으로 트리를 순회하면서, 해당 노드의 서브트리들이 t와 같은지 비교하는 sameTree()를 돌린다.
#sameTree()도 재귀 방식으로 돌아가는데, 해당노드의 값이 같은 것을 확인했으면, 다음으로 left child, right child의 값을 비교해 보고, 
#이 값들이 모두 True인 경우 True를 반환하고, else문에 걸려 하나라도 False가 있으면 False를 반환한다.(and이기 때문)





# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isSubtree(self, s: TreeNode, t: TreeNode) -> bool:
        self.identifier = []

        def dfs(s):
            if not s: return
            self.identifier.append(sameTree(s, t))
            if s.left: dfs(s.left)
            if s.right: dfs(s.right)
                
        #leetcode 100. Same Tree 
        def sameTree(p,q):
            if p and q:
                return p.val == q.val and sameTree(p.left, q.left) and sameTree(p.right, q.right)
            else:
                return p is q
    
        dfs(s)
        return any(self.identifier)
```
