Given two binary trees and imagine that when you put one of them to cover the other, some nodes of the two trees are overlapped while the others are not.

You need to merge them into a new binary tree. The merge rule is that if two nodes overlap, then sum node values up as the new value of the merged node. Otherwise, the NOT null node will be used as the node of new tree.


```python
#문제
#두개의 주어진 트리를 merge하라.(같은 위치의 수를 합쳐서)


class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution(object):
    def mergeTrees(self, t1, t2):                                      #가장 상단의 부모 tree부터 간다. 
        if not t1 and not t2: return None
        ans = TreeNode((t1.val if t1 else 0) + (t2.val if t2 else 0)) #tree1과 tree2의 값을더하는데, 한쪽의 값이 없다면 0으로 친다.
        ans.left = self.mergeTrees(t1 and t1.left, t2 and t2.left)    #왼쪽 child 부터 더해나간다. 재귀로 다시 처음부터 시작한다. left와 right의 값이 모두 없어져서 if not t1 and not t2: return None 조건이 만족할때까지.
        ans.right = self.mergeTrees(t1 and t1.right, t2 and t2.right) #ans.left가 모두 실행되면 ans.right가 실행된다. 실행원리는 위와 같다.
        return ans
```

```python
#Tree 연습1

#Inserting into a tree
class Node:

    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def insert(self, data):
    # Compare the new value with the parent node
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data

# Print the tree
    def PrintTree(self):
        if self.left:
            self.left.PrintTree()
        print(self.data),
        if self.right:
            self.right.PrintTree()

# Use the insert method to add nodes
root = Node(12)
root.insert(6)
root.insert(14)
root.insert(3)

root.PrintTree()
```

```python
#Tree 연습2

#참조 url
#https://onestep.tistory.com/42
#그림 5 참조


class Tree:
    def __init__(self, data, left_child = None, right_child = None):
        self.data = data
        self.left_child = left_child
        self.right_child = right_child

def preorder_traverse(tree): #전위 순회(Preorder Traverse) : 뿌리를 먼저 방문 한다.
    if tree == None: return
    print(tree.data)
    preorder_traverse(tree.left_child)
    preorder_traverse(tree.right_child)

def inorder_traverse(tree):  #중위 순회(Inorder Traverse) : 뿌리를 중간에 방문 한다.
    if tree == None: return
    inorder_traverse(tree.left_child)
    print(tree.data)
    inorder_traverse(tree.right_child)

def postorder_traverse(tree):  #후위 순회(Postorder Traverse) : 뿌리를 나중에 방문 한다.
    if tree == None: return
    postorder_traverse(tree.left_child)
    postorder_traverse(tree.right_child)
    print(tree.data)

levelq = []
def levelorder_traverse(tree):  #층별순회(Levelorder Traverse) : 레벨별로 방문 한다.
    global levelq
    levelq.append(tree)
    while len(levelq) != 0:
        # visit
        visit_node = levelq.pop(0)
        print(visit_node.data)
        # child put
        if visit_node.left_child != None:
            levelq.append(visit_node.left_child)
        if visit_node.right_child != None:
            levelq.append(visit_node.right_child)


root = None
def init_tree():
    global root
    # create leaf node
    leaf = []
    for i in range(6):
        leaf.append( Tree(i+1) )
    # create sub tree
    left_subtree = Tree(9, Tree(7, leaf[0], leaf[1]), Tree(8, leaf[2], leaf[3]))
    right_subtree = Tree(10, leaf[4], leaf[5])
    # create root
    root = Tree(11, left_subtree, right_subtree)

def Main():
    init_tree()
    print("< Preorder Traverse >") #전위 순회(Preorder Traverse) : 뿌리를 먼저 방문 한다.
    preorder_traverse(root)
    print()
    print("< Inorder Traverse >") #중위 순회(Inorder Traverse) : 뿌리를 중간에 방문 한다.
    inorder_traverse(root)
    print()
    print("< Postorder Traverse >") #후위 순회(Postorder Traverse) : 뿌리를 나중에 방문 한다.
    postorder_traverse(root)
    print()
    print("< Levelorder Traverse >") #층별순회(Levelorder Traverse) : 레벨별로 방문 한다.
    levelorder_traverse(root)
    print()

Main()
```
