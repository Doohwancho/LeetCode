```python
"""
Given a binary tree, each node has value 0 or 1.  Each root-to-leaf path represents a binary number starting with the most significant bit.  For example, if the path is 0 -> 1 -> 1 -> 0 -> 1, then this could represent 01101 in binary, which is 13.

For all leaves in the tree, consider the numbers represented by the path from the root to that leaf.

Return the sum of these numbers.

 

Example 1:



Input: [1,0,1,0,1,0,1]
Output: 22
Explanation: (100) + (101) + (110) + (111) = 4 + 5 + 6 + 7 = 22
"""

#문제

#1와 0으로 이루어진 binary tree가 있다. root부터 leaf까지의 각각의 tree path에 해당하는 노드의 값들은 2진수이다.
#모든 leaf의 이진수를 정수로 바꾼 값들의 합을 구하라.



#문제풀이

#dfs로 트리를 돌면서 해당 노드의 값들을 binary 변수에 string 형태로 붙여주고, 파라미터값으로 binary변수를 넘겨준다.
#if root.left or root.right: else:문으로, children이 없으면(leaf이면) leaf라는 리스트에 매 노드마다 해당 노드의 값들을 붙여준 binary를 넣어준다.
#리스트에 넣어진 string type의 binary number들은, int(binary number, 2)를 이용하여 정수로 바꿔주고, 리스트 값들의 총 합을 max()로 구한다.




# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def sumRootToLeaf(self, root: TreeNode) -> int:
        binary = ''
        leaf = []
        
        def dfs(root, binary):
            if not root: return
            binary += str(root.val)
            if root.left or root.right:
                if root.left: dfs(root.left, binary)
                if root.right: dfs(root.right, binary)
            else:
                leaf.append(binary)
        
        dfs(root, binary)
        return sum([int(x,2) for x in leaf])
```
