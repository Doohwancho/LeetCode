```python
"""
Given a binary tree and a sum, determine if the tree has a root-to-leaf path such that adding up all the values along the path equals the given sum.

Note: A leaf is a node with no children.

Example:

Given the below binary tree and sum = 22,

      5
     / \
    4   8
   /   / \
  11  13  4
 /  \      \
7    2      1
"""

#문제

#정수 sum이 주어지면, root~leaf의 총 합이 sum과 같은 것이 해당 트리안에 존재하면 True를 반환하라




#문제풀이

#depth first search 방식으로 순회하면서, 파라미터값으로 sum도 넘긴다.
#노드가 넘어가면, sum - 해당노드의 값(node.val)을 해준다.
#이렇게 계속 빼나가다가 leaf에 도착했을 때, sum이 0인경우 




# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def hasPathSum(self, root: TreeNode, sum: int) -> bool:
        self.identifier = False
        def dfs(root, sum):
            if not root: return
            sum = sum - root.val         #root에서부터 한 노드씩 내려갈때마다 해당 노드의 값(node.val)을 빼준다.
            print(sum, self.identifier)
            if root.left or root.right:  #left child, right child가 존재하지 않으면 leaf라는 뜻
                if root.left: dfs(root.left, sum)
                if root.right: dfs(root.right, sum)
            elif sum == 0:               #leaf에 도착했을 때, 계속 까내려가던 sum의 값이 0이면 True반환
                self.identifier = True    
            return self.identifier  
        
        return dfs(root,sum) if root else False
            
```
