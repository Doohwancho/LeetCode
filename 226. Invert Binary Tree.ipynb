```python
"""
Invert a binary tree.

Example:

Input:

     4
   /   \
  2     7
 / \   / \
1   3 6   9
Output:

     4
   /   \
  7     2
 / \   / \
9   6 3   1
"""

#문제
#주어진 tree를 좌우로 inverse 시켜라

#문제풀이
#재귀로 해당 루트가 존재하면, 해당 루트의 오른쪽 자식을 왼쪽자식으로, 왼쪽자식을 오른쪽 자식으로 넣어준다

# step01
#      4
#    /   \
#   2     7
#  / \   / \
# 1   3 6   9
    
# step02
#      4
#    /   \
#   7     2
#  / \   / \
# 6   9 1   3
    
# step03
#      4
#    /   \
#   7     2
#  / \   / \
# 9   6 3   1



# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def invertTree(self, root: TreeNode) -> TreeNode:
        if root:
            root.left, root.right = self.invertTree(root.right), self.invertTree(root.left)
            return root
```
