```python
"""
Given a binary tree, find the length of the longest path where each node in the path has the same value. This path may or may not pass through the root.

The length of path between two nodes is represented by the number of edges between them.

 

Example 1:

Input:

              5
             / \
            4   5
           / \   \
          1   1   5
Output: 2

 

Example 2:

Input:

              1
             / \
            4   5
           / \   \
          4   4   5
Output: 2
"""

#문제

#연속되게 같은 숫자가 이어지는 node들 중, 가장 긴 것이 몇번 이어지는지 반환하라
#예를들어, ex1의 경우, 같은 숫자로 이어지는 node들 중, 가장 긴 것은 5-5-5 이니까 5를 잇는 '-'가 2번있으므로, 2를 반환한다.


#문제 풀이

#Trial01 

#다음 노드로 넘어갈때마다, 이전노드의 값을 prev에 저장하면서, 자식으로 넘어갔을 때, 자식노드의 값과 prev의 값이 같다면 +1을 하는 방식으로 시도함.
#문제점1. 'longest edge' 파악 불가
#문제점2. left child leaf에서 right child leaf으로 건너뛸때 노드가 끊겨도 +1해줌

import collections, operator

class Solution:
    def longestUnivaluePath(self, root: TreeNode) -> int:
        edge = 0
        prev = None
        tracker = collections.defaultdict(int)
        
        def dfs(root, prev):
            if not root: return
            if prev == root.val: tracker[root.val] += 1
            prev = root.val
            if root.left: dfs(root.left, prev)
            if root.right: dfs(root.right, prev)
            
        
        dfs(root, prev)

        return max(tracker.items(), key=operator.itemgetter(1))[1] if tracker else 0
    
    
#Solution by yangshun(https://leetcode.com/problems/longest-univalue-path/discuss/108142/Python-Simple-to-Understand)

#접근 방법은 https://leetcode.com/problems/diameter-of-binary-tree/ 와 비슷
#다른점은, current node의 값이 child node의 값과 같지 않다면 0으로 초기화 해줌
#그러면서 longest 변수에는 max()함수를 이용해 가장 node의 길이가 큰것을 추적해줌



# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

import collections, operator

class Solution:
    def longestUnivaluePath(self, root: TreeNode) -> int:
        longest = [0]
        def traverse(node):
            if not node: return  
            left_len, right_len = traverse(node.left), traverse(node.right)
            left = (left_len + 1) if node.left and node.left.val == node.val else 0
            right = (right_len + 1) if node.right and node.right.val == node.val else 0
            longest[0] = max(longest[0], left + right)
            return max(left, right)
        traverse(root)
        return longest[0]
```
