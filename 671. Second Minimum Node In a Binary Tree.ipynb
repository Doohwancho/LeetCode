```python
"""
Given a non-empty special binary tree consisting of nodes with the non-negative value, where each node in this tree has exactly two or zero sub-node. If the node has two sub-nodes, then this node's value is the smaller value among its two sub-nodes.

Given such a binary tree, you need to output the second minimum value in the set made of all the nodes' value in the whole tree.

If no such second minimum value exists, output -1 instead.

Example 1:

Input: 
    2
   / \
  2   5
     / \
    5   7

Output: 5
Explanation: The smallest value is 2, the second smallest value is 5.
 

Example 2:

Input: 
    2
   / \
  2   2

Output: -1
Explanation: The smallest value is 2, but there isn't any second smallest value.
"""

#문제

#주어진 트리에서 두번째로 작은 수를 반환하라.
#두 자식중 더 작은수가 부모이고, 자식은 left,right 두개 모두 존재하거나 없다.
#만약 부모:2, left child:2, right child:2 처럼 트리의 모든 노드의 값이 같다면 -1를 반환하라.



#문제풀이

#bfs방식으로 트리를 순회하기 때문에 [root]를 bfs변수에 담아둔다.
#while 문을 돌리면서, 최상위에 있는 부모노드의 값(가장 작은 수)를 제외한 모든 수를 comprehension으로 구하고
#해당 값들의 최소값을 구하면, 가장 작은수가 아닌 다른 모든값들의 최소값이니, 모든 노드의 값들 중 두번째로 작은 값을 구할 수 있다.
#만약 모든 노드의 값이 최상위 노드의 값과 같아서, if node.val != root.val 이 부분의 조건이 맞지 않아 
#secondMin의 최소값이 float("inf")가 나오면, 모든 노드의 값이 같으므로, -1를 반환해 준다.



# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution: # by cenkay
    def findSecondMinimumValue(self, root: TreeNode) -> int:
        secondMin, bfs = float('inf'), [root]
        while bfs:
            secondMin=min([secondMin]+[node. val for node in bfs if node.val != root.val])
            bfs = [child for node in bfs for child in (node.left, node.right) if child]

        return secondMin if secondMin != float('inf') else -1
```
